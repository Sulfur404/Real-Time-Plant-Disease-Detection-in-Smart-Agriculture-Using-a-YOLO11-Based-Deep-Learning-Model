```python
from ultralytics import YOLO
import time


model = YOLO("/kaggle/input/yolo11x-plant-disease-model/best.pt")

import torch
dummy = torch.randn(1, 3, 640, 640).cuda()

# Warm-upGPU 
for _ in range(5):
    _ = model(dummy)

# Measure FPS
start = time.time()
iters = 50

for _ in range(iters):
    _ = model(dummy)

end = time.time()

fps = iters / (end - start)
print(f"GPU Inference Speed: {fps:.2f} FPS")
```

```python
from ultralytics import YOLO
import time
import torch

model = YOLO("/kaggle/input/yolo11x-plant-disease-model/best.pt")

dummy = torch.randn(1, 3, 640, 640)

# Warm-up 
for _ in range(2):
    _ = model(dummy)

iters = 20
start = time.time()

for _ in range(iters):
    _ = model(dummy)

end = time.time()

fps_cpu = iters / (end - start)
print(f"CPU Inference Speed: {fps_cpu:.2f} FPS")
```
